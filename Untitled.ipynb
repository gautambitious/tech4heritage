```python
import efficientnet.tfkeras
from tensorflow.keras.models import load_model
```

```python
model = load_model('patch_detection.h5',compile=False)
```

```python
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np
```

```python
patched_img = np.array(Image.open('check.png'))
```

```python
patched_img.shape
```

```python
patched_img = patched_img.reshape(1,256,256,3)
p = model.predict(patched_img)
```

```python
plt.imshow((p.reshape(256,256)*256).astype(np.uint8))
```

```python
final = Image.fromarray((p.reshape(256,256)*256).astype(np.uint8))
```

```python
final.save('../check.png')
```
