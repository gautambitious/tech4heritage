```python
import efficientnet.tfkeras
from tensorflow.keras.models import load_model
```

```python
model = load_model('patch_detection.h5',compile=False)
```

```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

```python
image = cv2.imread('check.png', cv2.IMREAD_UNCHANGED)
```

```python
image.shape
```

```python
image.resize(256,256,3)
```

```python
patched_img = image.reshape(1,256,256,3)
p = model.predict(patched_img)
```

```python
plt.imshow((p.reshape(256,256)*256).astype(np.uint8))
```

```python
final = (p.reshape(256,256)*256).astype(np.uint8)
```

```python
final
```

```python
cv2.imwrite('bruh.png', final)
```
